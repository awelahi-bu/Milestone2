```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python



# Load datasets
social_media_data = pd.read_csv('Social_Media_Advertising.csv')
supply_chain_data = pd.read_csv('supply_chain_data.csv')

# Social Media Dataset Graphs
# Conversion Rate distribution histogram
plt.figure(figsize=(10, 6))
sns.histplot(social_media_data['Conversion_Rate'], bins=30, kde=True)
plt.title('Conversion Rate Distribution - Social Media Advertising')
plt.xlabel('Conversion Rate')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# ROI distribution boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=social_media_data['ROI'])
plt.title('ROI Distribution - Social Media Advertising')
plt.xlabel('ROI')
plt.grid(True)
plt.show()

# Channel Used vs Engagement Score bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Channel_Used', y='Engagement_Score', data=social_media_data)
plt.title('Channel Used vs Engagement Score')
plt.xlabel('Channel Used')
plt.ylabel('Average Engagement Score')
plt.grid(True)
plt.show()

# Supply Chain Dataset Graphs
# Scatter plot comparing Revenue and Costs
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Revenue generated', y='Costs', data=supply_chain_data)
plt.title('Revenue vs Costs - Supply Chain')
plt.xlabel('Revenue Generated')
plt.ylabel('Costs')
plt.grid(True)
plt.show()

# Product Availability Histogram
plt.figure(figsize=(12, 6))
sns.histplot(supply_chain_data['Availability'], bins=30, kde=True)
plt.title('Product Availability')
plt.xlabel('Availability')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# Histogram illustrating Lead Times distributions
plt.figure(figsize=(10, 6))
sns.histplot(supply_chain_data['Lead times'], bins=30, kde=True)
plt.title('Lead Time Distribution')
plt.xlabel('Lead Time')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Heatmap indicating correlation between numerical variables in Supply Chain data
plt.figure(figsize=(14, 10))
sns.heatmap(supply_chain_data.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Supply Chain Data Correlations')
plt.show()

# Select relevant numeric columns from each dataset
ad_metrics = social_media_data[['Conversion_Rate', 'ROI', 'Clicks', 'Impressions', 'Engagement_Score']]
supply_chain_metrics = supply_chain_data[['Revenue generated', 'Costs', 'Lead times', 'Availability', 'Number of products sold']]

# Concatenate datasets horizontally (assuming rows correspond to similar records or campaigns)
combined_metrics = pd.concat([ad_metrics.reset_index(drop=True), supply_chain_metrics.reset_index(drop=True)], axis=1)

# Calculate correlation matrix
corr_matrix = combined_metrics.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Advertising Metrics and Supply Chain Metrics')
plt.show()
```
